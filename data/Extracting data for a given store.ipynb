```python
'''Trying to find out a store that's quite big and have potential to have a big number
of stores to rely data on. '''

stores_csv = pd.read_csv('stores.csv')
stores_csv.head()
```

```python
#Getting the row (i.e. index) of a store that's very big in size
index = stores_csv['Size'].idxmin()
index
```

```python
'''The row representing the store with the biggest size'''
store_row = stores_csv.values[index]
print(store_row)
STORE_NUMBER = store_row[0]
STORE_NUMBER
```

```python
#Count rows of features based on the store number
# Also the records represent weeks so it makes sense that it's 182 weeks rather than 182 days
features_csv = pd.read_csv('features.csv')
store_row = features_csv.Store
len([store_id for store_id in store_row if store_id == STORE_NUMBER])
```

```python
# Extracting a set of data examples for the store number
features_set = features_csv[features_csv.Store == STORE_NUMBER]
# I noticed due to the features_set.count that there were some inconsistencies with unemployment dimension
features_set[pd.isnull(features_set.Unemployment)]
# Update: However, since discovery, I realised they fall out of range for the historical data, so I should not worry since
# I need to extract the features set for that particular range.
```

```python
# Checking on the historical data of weekly sales
store_csv = pd.read_csv('train.csv')
store_set = store_csv[store_csv.Store == STORE_NUMBER]
store_set
```

```python
# So based on the found period range the historical data (i.e. store_set),
# I will extract the features for the given period range
features_set = features_set[features_set.Date <= '2012-09-14']
features_set.tail()
```

```python
# Writing extracted data to the files
store_set.to_csv('extracted_data/train.csv')
features_set.to_csv('extracted_data/features.csv')
```
